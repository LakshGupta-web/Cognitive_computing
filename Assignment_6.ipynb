```python
import numpy as np
import matplotlib.pyplot as plt
M=float(input('Enter a number: '))
x=np.linspace(-10,10,100)
y1=M*x**2
y2=M*np.sin(x)
plt.plot(x,y1,'r--',label='y=M*x^2')
plt.plot(x,y2,'b-',label='y=M*sin(x)')
plt.legend()
plt.grid()
plt.title('Mathematical Functions')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Geography'],
    'Score': [85, 90, 78, 88, 92]
}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x='Subject', y='Score', data=df, palette='viridis')
for index, row in df.iterrows():
    bar_plot.text(index, row['Score'], str(row['Score']), color='black', ha="center")
plt.title('Scores by Subject', fontsize=16)
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Scores', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(12345)
data = np.random.randn(50)
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].plot(np.cumsum(data), label='Cumulative Sum', color='blue')
axes[0].set_title('Cumulative Sum of Data', fontsize=14)
axes[0].set_xlabel('Index', fontsize=12)
axes[0].set_ylabel('Cumulative Sum', fontsize=12)
axes[0].grid(True)
axes[0].legend()
axes[1].scatter(range(len(data)), data, label='Random Data', color='green', alpha=0.7)
axes[1].set_title('Scatter Plot with Random Noise', fontsize=14)
axes[1].set_xlabel('Index', fontsize=12)
axes[1].set_ylabel('Values', fontsize=12)
axes[1].grid(True)
axes[1].legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
url = 'https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv'
df = pd.read_csv(url)
df['month_number'] = df['month_number'].astype(str)
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['month_number'], y=df['total_profit'], marker='o', color='blue')
plt.title('Total Profit of All Months', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Profit', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['month_number'], y=df['facecream'], label='Facecream', marker='o')
sns.lineplot(x=df['month_number'], y=df['facewash'], label='Facewash', marker='o')
sns.lineplot(x=df['month_number'], y=df['toothpaste'], label='Toothpaste', marker='o')
sns.lineplot(x=df['month_number'], y=df['bathingsoap'], label='Bathingsoap', marker='o')
sns.lineplot(x=df['month_number'], y=df['shampoo'], label='Shampoo', marker='o')
sns.lineplot(x=df['month_number'], y=df['moisturizer'], label='Moisturizer', marker='o')
plt.title('Sales Data for All Products', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.legend(title='Products')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
df_melted = df.melt(id_vars=['month_number'], var_name='Feature', value_name='Value')
plt.figure(figsize=(12, 6))
sns.barplot(x='Feature', y='Value', hue='month_number', data=df_melted)
plt.title('Bar Chart for All Features/Attributes', fontsize=16)
plt.xlabel('Feature', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
